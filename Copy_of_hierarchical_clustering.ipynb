# Hierarchical Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```
import scipy.cluster.hierarchy as sch
# Linkage takes two arguments, first one is matrix of features and second is technique
# ward technique is basically minimizing the variance of the observations inside the cluster

dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers(Observation Points)')
plt.ylabel('Euclidean Distances')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```
from sklearn.cluster import AgglomerativeClustering
# affinity means the type of distance you want to compute
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```
print(y_hc)
```

```
# Let's try with 3 clusers
from sklearn.cluster import AgglomerativeClustering
# affinity means the type of distance you want to compute
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

## Visualising the clusters

```
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 25, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 25, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 25, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 25, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 25, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (x$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```
# With 3 clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 25, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 25, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 25, c = 'green', label = 'Cluster 3')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (x$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
